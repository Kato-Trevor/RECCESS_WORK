# Series

### 1.	Write a Pandas program to compare the elements of two Pandas series

```python
import pandas as pd
import numpy as np

# sample series
series1 = pd.Series([4, 65, 436, 3, 9])
series2 = pd.Series([7, 0, 3, 897, 9])

# compare the elements
comparison = series1 == series2

# comparison results
comparison_df = pd.DataFrame({'Series 1': series1, 'Series 2': series2, 'Comparison': comparison})

# display
print(comparison_df)
```

### 2. Write a Pandas program to add, subtract, multiply and divide two Pandas series

```python
series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])

# Addition
addition = series1 + series2

# Subtraction
subtraction = series1 - series2

# Multiplication
multiplication = series1 * series2

# Division
division = series1 / series2

# Create a new DataFrame to display the results
result_df = pd.DataFrame({'Series 1': series1, 'Series 2': series2, 'Addition': addition, 'Subtraction': subtraction,
                          'Multiplication': multiplication, 'Division': division})

# Display the result DataFrame
print(result_df)
```

### 5.	Write a Pandas program to display the most frequent value in the given series and replace everything else as ‘Other’ in the series

```python
df_series = pd.read_csv('C:/Users/USER/RECESS/datasets/AfricaCupofNationsMatches.csv') 
df_series['HomeTeamGoals']
```

```python
# Calculate the value counts for the column
value_counts = df_series['HomeTeamGoals'].value_counts()
# Get the most frequent value
most_frequent_value = value_counts.index[0]
most_frequent_value

# Replace everything else as 'Other' in the series
df_series['HomeTeamGoals'] = df_series['HomeTeamGoals'].apply(lambda x: most_frequent_value if x == most_frequent_value else 'Other')

df_series
```
