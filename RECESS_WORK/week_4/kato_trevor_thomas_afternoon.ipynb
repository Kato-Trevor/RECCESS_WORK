```python
#making a series from a dictionary
my_calendar = {
    "Monday": "01/02/2023",
    "Tuesday": "02/02/2023",
    "Wednesday": "03/02/2023",
    "Thursday": "04/02/2023",
    "Friday": "05/02/2023"
}

# Create a Pandas Series from the dictionary
series = pd.Series(my_calendar)

# Print the Pandas Series
print(series)
```

```python
my_new_series = series.loc[["Monday", "Tuesday"]]

# Print the Pandas Series
print(my_new_series)
```
